```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
df1 = pd.read_csv("bengaluru_house_prices.csv")
df1.head()
```

```python
df2 = df1.drop(["availability","area_type",'society','balcony'],axis = 'columns')
```

```python
df3=df2.dropna()
```

```python
#size contains string BHK,bedroom working on which is not possible.Hence it is to be removed .create a bhk column to store number
df3['bhk'] = df3['size'].apply(lambda x: x.split(' ')[0])
```

```python
df4 = df3.drop('size',axis = 'columns')

#bhk column needs to be converted into float 
df4.bhk = df4.bhk.apply(lambda x: float(x))
```

```python
#function to check if data is float or not
def is_float(x) :
    try :
        float(x)
    except:
        return False
    return True
```

```python
#in df4 not all total sqft datasets are float . given below is range. which is data error and needs to be corrected
df4[~df4['total_sqft'].apply(lambda x: is_float(x))]
```

```python
#function to take the average of range 
def rng(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None   
df4.total_sqft=df4.total_sqft.apply(rng)
```

```python
df4 = df4[df4['total_sqft'].notnull()]
df4.loc[30]

#as you can see total sqft is mean of the range the range 2100-2850 
```

```python
#converting price to float
df4.price.apply(lambda x:float(x))
```

```python
#calculating price per sqft
df4['price_per_sqft']=(df4['price'])*100000/(df4['total_sqft'])
```

```python
df4
```

```python
#there are many locations whose occurence is single time
location_stats = df4.location.value_counts(ascending=False)
location_stats
```

```python
len(location_stats[location_stats>10])
```

```python
location_less_than_10 = location_stats[location_stats<=10]
```

```python
#I have set 10 as limit ,below which location wil be classified as others
df4.location = df4.location.apply(lambda x : 'others' if x in location_less_than_10 else x)
```

```python
df4.head(10)
```

```python
#considering minimum size of a bedroom(including bathroom) as 300sqft anyvalue below that is outlier and needs to be removed
df5  = df4[~(df4.total_sqft/df4.bhk<300)]
df5.shape
df5.price_per_sqft.describe()
```

```python
#as describe shows there is a huge difference in min and max .our data should be in between mean+-std for more accurate prediction
#so we need to remove price per sqft(pps) outlier
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out
df6 = remove_pps_outliers(df5)
df6.shape
```

```python
#in standard apartment the bathroom are equal to bedroom or bedroom+2 everything else can be considered as an  outlier and
#can be removed safely.
df7 = df6[df6.bath<df6.bhk+2] 
df7
```

```python
#we need to convert location into binary using dummy variables
dummies = pd.get_dummies(df7.location)
from sklearn.model_selection import train_test_split
dummies
```

```python
#for accuracy purpose we drop others column .as it contains places with threshold <10
df8 = pd.concat([df7,dummies.drop('others',axis='columns')],axis='columns')
df8
```

```python
#we need to split model in dependent and independent variables in order to train it.
#X here is independent 

X = df8.drop(['price','location','price_per_sqft'],axis='columns')
```

```python
#y is dependent
y=df8.price
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
lr.score(X_test,y_test)
```

```python
#we import libraries and functions . We define best_model function to get score for each model.Dictionary algorithms contain 
#model and parameters

from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def best_model(X,y):
    
    algorithms={
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {
                'normalize': [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores =[]
    cv = ShuffleSplit(n_splits=5 , test_size= 0.2 , random_state=0)
    for algoname , config in algorithms.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X,y)
        print(gs.best_score_)
        gs.best_params_
        scores.append(
        {
            'model':algoname,
            'best_score':gs.best_score_,
            'best_params': gs.best_params_
        }
        )
    print(scores)    
    return pd.DataFrame(scores,columns=['model','best_score','best_params'])
best_model(X,y)
```

```python
#predict price function to predict values.we have used linear regression model as it gives the best score.
def predict_price(location,sqft,bath,bhk):    
    loc_index = np.where(X.columns==location)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk
    if loc_index >= 0:
        x[loc_index] = 1

    return lr.predict([x])[0]
```

```python
predict_price('1st Phase JP Nagar',1000, 2, 2)
```
